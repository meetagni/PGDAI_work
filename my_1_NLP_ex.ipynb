```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
text = "This is a sample sentence for NLP programming." 
# Tokenize the text 
tokens = nltk.word_tokenize(text) 
# Perform parts of speech tagging 
tagged = nltk.pos_tag(tokens) 
# Print the result 
print(tagged)
```
