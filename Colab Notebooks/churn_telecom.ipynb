MITIGATING SKEWNESS OF DATA

```python
#method 1: undersampling
```

```python
count_class_count0, count_class_1=df2.Churn.value_counts()
```

```python
df_class_0=df2[df2['Churn']==0]
df_class_1=df2[df2['Churn']==1]
```

```python
df_class_0.shape
```

```python
df_class_0_under=df_class_0.sample(count_class_1)
```

```python
df_class_0_under.shape
```

```python
df_class_0_under.Churn.values
```

```python
df_final_under= pd.concat([df_class_0_under, df_class_1], axis=0)
df_final_under.shape
```

```python
df_final_under.Churn.value_counts()
```

```python
X=df_final_under.drop('Churn', axis=1)
y=df_final_under['Churn']
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2, random_state=100, stratify=y)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
def ANN(X_train, X_test, y_train, y_test, loss):
  model=keras.Sequential([
    keras.layers.Dense(26,input_shape=(26,), activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
    ])

  model.compile(optimizer='adam',
              loss=loss,
              metrics=['accuracy']
              )
  model.fit(X_train,y_train, epochs=100)
  print(model.evaluate(X_test,y_test))
  y_preds=model.predict(X_test)
  y_preds=np.round(y_preds)
  print("Classification report:\n",classification_report(y_test,y_preds))
  return y_preds
```

```python
y_pred=ANN(X_train, X_test, y_train, y_test, 'binary_crossentropy')
```

Method 4:
Ensemble with undersampling

```python
##0===4130
##1===1869
df3=X_train.copy()
df3['Churn']=y_train
df3.head()
```

```python
df3_class0=df3[df3['Churn']==0]
```

```python
df3_class1=df3[df3['Churn']==1]
```

```python
def get_train_batch(df_major, df_minor, start, end):
  df_train=pd.concat([df_major[start:end],df_minor],axis=0)
  X_train=df_train.drop('Churn',axis=1)
  y_train=df_train['Churn']
  return X_train,y_train
```

```python
X_train, y_train=get_train_batch(df3_class0,df3_class1,0,1495)
ypred1=ANN(X_train,X_test,y_train,y_test,'binary_crossentropy')
```

```python
X_train, y_train=get_train_batch(df3_class0,df3_class1,1496,2990)
ypred2=ANN(X_train,X_test,y_train,y_test,'binary_crossentropy')
```

```python
X_train, y_train=get_train_batch(df3_class0,df3_class1,2990,4485)
ypred3=ANN(X_train,X_test,y_train,y_test,'binary_crossentropy')
```

```python
y_pred_final=ypred1.copy()
for i in range(len(ypred1)):
  n_ones=ypred1[i]+ypred2[i]+ypred3[i]
  if n_ones>1:
    y_pred_final[i]=1
  else:
    y_pred_final[i]=0
    
```

```python
print(classification_report(y_test,y_pred_final))
```
