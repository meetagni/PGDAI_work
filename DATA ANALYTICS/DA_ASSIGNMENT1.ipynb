```python
import numpy as np
import pandas as pd
car=pd.read_csv(r"C:\Users\Meet\Downloads\car.csv")
car.head()
```

### 1. Most expensive car for each car company make.

```python
car.groupby(['Make'], as_index=False).MSRP.max()
```

### 2. Most expensive convertible vehicle for each company.

```python
car[car['Vehicle Style'] == 'Convertible'].groupby(['Make'] , as_index=False).MSRP.max()
```

### 3. List all the cars which are automatic.

```python
car[car['Transmission Type'] == 'AUTOMATIC'].groupby(['Make','Model']).first()
```

### 4. Premium cars of each car company

```python
car[car['Engine Fuel Type'] == ('premium unleaded (required)' or 'premium unleaded (recommended)')].groupby(['Make','Model']).first()
```

### 5. Average price of cars each car company make.

```python
car.groupby(['Make']).MSRP.mean()
```
