```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
start=pd.to_datetime('2012-12-20')
end=pd.to_datetime('2022-12-20')
```

```python
df = yf.download("GOOG AMZN TSLA", start=start, end=end,group_by="ticker") 
df
```

```python
df.head()
```

```python
df.GOOG['Adj Close'].plot()
df.AMZN['Adj Close'].plot()
df.TSLA['Adj Close'].plot()
```

```python
stock_data=pd.concat([df.AMZN['Adj Close'],df.GOOG['Adj Close'],df.TSLA['Adj Close']],axis=1)
stock_data.head()
```

```python
logReturns = np.log(stock_data/stock_data.shift(1))

#daily average return
logReturns.mean()
```

```python
#annual average return
logReturns.mean() * 250
```

```python
#correrlation
stock_data.corr()
```

```python
Sharpe_Ratio = logReturns.mean().mean() / logReturns.mean().std()
Sharpe_Ratio
```

```python
#Daily standard mean
logReturns.std()
```

```python
expectedReturn = []
standardDeviation = []
weightList0 = []
weightList1 = []
weightList2 = []
numberOfStocks = 3

# Running simulations for finding optimum weights
for i in range(100):
    weights = np.random.random(numberOfStocks)
    weights = weights/ weights.sum()
    weightList0.append(weights[0])
    weightList1.append(weights[1])
    weightList2.append(weights[2])
    expectedReturn.append((weights * logReturns.mean()).sum() * 250)
    standardDeviation.append(np.sqrt(np.dot(weights.T, np.dot(logReturns.cov() * 250, weights))))

#Converting lists into arrays
weightList0 = np.array(weightList0) #Weights for AMZN
weightList1 = np.array(weightList1) #Weights for GOOG
weightList2 = np.array(weightList2) #Weights for TSLA
expectedReturn = np.array(expectedReturn) 
standardDeviation = np.array(standardDeviation)

#Creating dataframe
df2 = pd.DataFrame({"Weight of Amazon": weightList0, "Weight of Google": weightList1, "Weight of Tesla": weightList2, "Expected Return": expectedReturn, "Standard deviation": standardDeviation})
df2.head()
```

```python
plt.figure(figsize=(14, 10), dpi=80)
plt.scatter(df2["Standard deviation"], df2["Expected Return"])
plt.xlabel("Standard deviation")
plt.ylabel("Expected return (in %)")
plt.show()
```
